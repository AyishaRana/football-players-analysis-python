## Checked Goal Contributions
added goals + assists to measure each player’s total contribution.
- Top scorers: Osimhen, Kane
- Top assister: Randal Kolo Muani
- 6 players with 20+ contributions

```python
goals_assists=df.groupby('Player Name')[['Goals','Assists']].sum()
goals_assists=goals_assists.sort_values(by='Goals',ascending=False)
goals_assists['Total contribution']=goals_assists['Goals'] + goals_assists['Assists']
goals_assists
```

## Looked at Playing Style and Strengths
compared other match statistics to see what type of striker each player is:
- Shots per game: Osimhen, Mitrovic
- Passing accuracy: David, Ramos, Thuram, Højlund
- Aerial duels: Mitrovic, Toney
- Ratings: Strong for multiple players


```python
avg=df.groupby('Player Name')[['Shots','PS%','AerialsWon','Rating']].mean().reset_index().round(0)
avg.columns=['Player Name', 'Avg Shots', 'Avg PS%', 'Avg  AerialsWon', 'Avg Rating']
avg=avg.sort_values(by=['Avg Shots', 'Avg PS%', 'Avg  AerialsWon', 'Avg Rating'],ascending=False).reset_index(drop=True)
avg
```

## Consistency Check
checked how stable their ratings are across matches (using averages, medians, and variation).

- Harry Kane and Dusan Vlahovic stood out as a very consistent performer

```python
summary = df.groupby('Player Name')['Rating'].agg(mean='mean', median='median', mode=lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan).round(0).reset_index()
summary.columns=['Player Name','mean Rating','median Rating','mode Rating']
summary
```

```python
range=df.groupby('Player Name')['Rating'].agg(max='max',min='min',range=lambda x: x.max()-x.min()).round(0).reset_index()
range.columns=['Player Name','Max of Rating','Min of Rating','Range']
range=range.sort_values(by='Max of Rating',ascending=False).reset_index(drop=True)
range
```

```python
iqr=df.groupby('Player Name')['Rating'].agg(Q1=lambda x: x.quantile(0.25),Q3=lambda x: x.quantile(0.75),IQR=lambda x: x.quantile(0.75)-x.quantile(0.25)).round(0)
iqr
```

```python
var=df.groupby('Player Name')['Rating'].agg(variance='var',std_dev='std').round(2).reset_index()
var.columns=['Player Name','Var. of Rating','StdDev. of Rating']
var=var.sort_values(by=['Var. of Rating','StdDev. of Rating']).reset_index(drop=True)
var
```
